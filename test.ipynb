```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
import accelerate
```

```python
model = AutoModelForCausalLM.from_pretrained("facebook/opt-1.3b", torch_dtype=torch.float32, cache_dir="./cache/", device_map="auto")
```

```python
sentence = "Which properties does a mathematical norm have to fulfill?"
tokenizer = AutoTokenizer.from_pretrained("facebook/opt-125m", cache_dir="./cache/")
input_ids = tokenizer.encode(sentence, return_tensors="pt").to(device=model.device)
```

```python
output = model.generate(input_ids, max_length=50, do_sample=True, top_p=0.95, top_k=60)
```

```python
print(tokenizer.decode(output[0], skip_special_tokens=True))
```
